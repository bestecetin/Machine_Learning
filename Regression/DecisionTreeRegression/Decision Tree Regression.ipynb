```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df=pd.read_csv("/Users/bestecetin/Desktop/machine_learning/courses/machinelearning/Regression/DecisionTreeRegression/data/decision+tree+regression+dataset .csv",sep=";",header=None)
```

- header = None yapmazsak ilk satırdaki value'ları header olarak yazıyordu biz 0 ve 1. index olarak gözükmesini sağladık

veri setimiz tribündeki levellerin kaç tl ücrete denk geldiğini tutuyor.

```python
x = df.iloc[:,0].values.reshape(-1,1) #koltuk level'i
y = df.iloc[:,1].values.reshape(-1,1) #ücret 
```

```python
#decision tree regression
from sklearn.tree import DecisionTreeRegressor
tree_reg=DecisionTreeRegressor()
tree_reg.fit(x,y)
```

```python
df.head()
```

```python
tree_reg.predict([[6]]) 
```

- Burada 6 verince 40'ı predict etmesini bekleriz ki öyle de oldu

```python
y_pred=tree_reg.predict(x)
```

```python
#visualize
plt.scatter(x,y, color="red")
plt.plot(x,y_pred,color="green")
plt.xlabel("tribun level")
plt.ylabel("ucret")
plt.show()
```

- Burada x değerlerini predict ettim x değerleri 1'den 10'a kadar olan değerlerdi. 
- Her bir leaf'de tahmin edilen değerlerimin hepsi aynı olmalıydı,
- Bu yüzden her leaf'de düz bir line bekliyorum.
- Bunu çözmek için büyük bir aralıkta prediction yapmamız gerekiyor. 

```python
x_=np.arange(min(x),max(x),0.01).reshape(-1,1)
y_pred=tree_reg.predict(x_)
```

- x_ değerinde minimum x değerinden maximum x değer,ne kadar 0.01 aralıklarla değer oluşturduk

```python
x_
```

```python
#visualize
plt.scatter(x,y, color = "red")
plt.plot(x_,y_pred,color = "green")
plt.xlabel("tribun level")
plt.ylabel("ucret")
plt.show()
```

- Bir leaf'den diğer leaf'e geçmediği sürece aynı değerde düz bir line
- leaf'ler arası geçişlerde split işlemi yapıldıgında aşağı doğru inen düz line'ı görebiliyorum.
- Şekilde görüldüğü gibi belli aralıklarda yani split’lerde y değerleri sabit kalmaktadır.
